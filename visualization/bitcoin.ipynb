```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import yfinance as yf

# Install if you don't have it

# Get Bitcoin data
btc = yf.download('BTC-USD', 
                  start='2019-01-01', 
                  end='2019-12-31')

btc.head()
```

```python
# Create figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Plot the closing price
ax.plot(btc.index, btc['Close'], color='#F7931A', linewidth=2, label='BTC-USD')

# Customize the plot
ax.set_title('Bitcoin Price Evolution in 2019', fontsize=14, pad=20)
ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Price (USD)', fontsize=12)

# Add grid
ax.grid(True, linestyle='--', alpha=0.7)

# Format y-axis to show thousands with $ symbol
ax.yaxis.set_major_formatter(lambda x, p: f'${x:,.0f}')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add legend
ax.legend()


# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show plot
plt.show()
```
